### 3. 다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)

```python
class Counter:
    def __init__(self, count, step=1):
        self.count = count
        self.step = step
    
    def __str__(self):
        return "[Count(step:{0})]{1}".format(self.step, self.count)
    
    def incr(self):
        self.count += self.step
    
    def __call__(self):
        self.incr()
        
    def __add__(self, num):
        newCount=self.count+num
        return Counter(newCount, self.step)
    
    def __sub__(self, num):
        newCount=self.count-num
        return Counter(newCount, self.step)
     
    def __gt__(self, num):
        return self.count > num
    
    def __lt__(self, num):
        return self.count < num
    
    def __eq__(self, num):
        return self.count == num
```

```python
c = Counter(10)
d = Counter(10, 2)
```

### 4. 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

__init__(): MySet클래스의 생성자로 리스트 l을 전달 받고 self에 그대로 추가하고 eliminate_duplicate() 함수를 통해 중복을 제거한다.  
  
__str__(self): 객체에 대한 정보를 문자열 타입으로 반환해주는 역할을 한다.  
1. 일단 "MySet: {"로 초기화된 문자열객체 result를 만들고  
2. for문으로 self에 요소들에 접근하며 문자열 타입으로 형변환 한뒤  
3. +연산자를 통해 result에 문자열 뒤에 " ,"와 함께 추가해준다. 이렇게 하면 result문자열에 self의 모든요소들이 들어가게 된다.
4. 반복문을 종료하고 슬라이싱을 활용해 맨 마지막에 찍힌 " ,"를 제거해 주고 "}"을 추가한다.  
5. 마지막으로 result를 반환한다.

elimicate_duplicate(): 요소들의 중복을 제거하는 역할을한다.   
1. s라는 리스트객체를 만든다. 
2. l의 요소들을 for문으로 접근한다.
3. 리스트 s에 해당 요소가 없으면 그 요소를 s에 추가한다. 이렇게 함으로써 s에는 동일한 값이 들어올 수 없다.
4. l을 비워주고 l에 리스트 s의 요소들을 for문을 통해 똑같이 추가해 준다.
5. 결과적으로 elimicate_duplicate()의 매개변수로 전달된 l은 중복이 모두 제거되게 된다.

### 5. 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result
   
    """
    | 연산자로 합집합을 반환하는 코드이다. 
    
    1. result라는 Myset객체를 만들고 self에 있는 요소들은 그대로 추가해준다.
    2. other에 요소들은 추가할때는 이미 result에 있는 요소인지 확인하고 추가한다.
    3. result 객체를 반환해준다.
    """
    def __or__(self, other):
        result=MySet([])
        for e in self:        
            result.append(e)
        for e in other:
            if e not in result:
                result.append(e)
        return result

    
    """
    & 연산자로 교집합을 반환하는 코드이다. 
    
    1. result라는 Myset객체를 만든다.
    2. other의 요소들을 접근하며 in 을 통해 만약 other의 요소가 self에 있으면 result에 추가한다.
    3. result 객체를 반환해준다.
    """
    def __and__(self, other):
        result=MySet([])
 
        for e in other:
            if e in self:
                result.append(e)
        return result
     
        
    """
    - 연산자로 차집합을 반환하는 코드이다. 
    
    1. result라는 Myset객체를 만든다.
    2. self의 요소들을 접근하며 self의 요소가 other에 없으면 result에 추가한다.
    3. result 객체를 반환해준다.
    """
    def __sub__(self, other): 
        result=MySet([])
        
        for e in self:
            if e not in other:
                result.append(e)
                
        return result


    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
    
      
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
    
```

```python
import requests
import string
class SearchEngine:
    
    
    #생성자이다. url을 가변인자로 받는다.
    #3개의 종합적인 단어의 빈도수, 사이트목록, 사이트마다 단어의 목록을 유지한다.
    def __init__(self, *url):
            self.totalFreq=dict()
            self.urlList=[]
            self.wordsFreq=[]
            for i in url:
                self.urlList.append(i)
                self.wordsFreq.append(SearchEngine.__wordsFrequency(i))
            
            for d in self.wordsFreq:
                for word, value in d.items():
                    if word in self.totalFreq:
                        self.totalFreq[word]+=value
                    else:
                        self.totalFreq[word]=value
                
            
    #url을 추가하는 메소드이다. url을 추가함에 따라 멤버변수에도 변화를 줘야한다.
    #url을 url목록에 추가하고, 사이트단어목록에 추가하고, 종합단어목록에 사이트의 단어를 추가한다.             
    def addUrl(self, url):
        self.urlList.append(url)
        self.wordsFreq.append(SearchEngine.__wordsFrequency(url))   

        for word, value in self.wordsFreq[len(self.wordsFreq)-1].items():
            if word in self.totalFreq:
                self.totalFreq[word]+=value
            else:
                self.totalFreq[word]=value
    
    #url을 제거하는 메소드이다. url을 제거함에 따라 멤버변수에도 변화를 줘야한다.
    #url을 url목록에서 제거하고 그 url의 단어목록도 제거하고, 종합단어목록도 제가한 url을 빼고 다시 초기화 해준다.    
    def removeUrl(self, url):
        if url not in self.urlList:
            print("해당 url은 존재하지 않습니다.")
            return
        
        removeIndex=self.urlList.index(url)
        
        self.wordsFreq.pop(removeIndex)       
        self.urlList.remove(url)
        
        self.totalFreq.clear()
        for d in self.wordsFreq:
            for word, value in d.items():
                if word in self.totalFreq:
                    self.totalFreq[word]+=value
                else:
                    self.totalFreq[word]=value
     
    #url의 목록을 출력해준다.
    def  listUrls(self):
        for i in self.urlList:
            print(i)
       
    #멤버변수를 통해 사이트의 종합 단어 빈도수를 출력해준다.
    def getWordsFrequency(self):
        return self.totalFreq
    
    #종합 단어 빈도수를 유지하는 멤버변수 totalFreq를 활용해 빈도수가 가장많은 단어를 출력해준다.
    #사전에서 가장 큰 value가 무엇인지 찾고 그 value에 해당하는 key들을 찾아 사전으로 만든뒤 반환해준다.
    #이렇게 하면 공동 1위가 있어도 같이 반환이 가능하다.
    def getMaxFreqencyWords(self):
        maxFrequencyWords=dict()
        
        l=list(self.totalFreq.values())
        l.sort(reverse=True)
        max_value=l[0]
        for key, value in self.totalFreq.items():
            if value==max_value:
                maxFrequencyWords[key]=value
        
        return maxFrequencyWords
    
    #단어를 검색하여 가장 유사도가 높은 url을 출력해주는 메소드이다.
    #저번에 만든 함수를 통해 만들었는데 유사도를 구하고 가장큰 유사도를 가지는 url을 출력하게 하였다.
    #만약 유사도가 같은 사이트가 있어도 같이 출력된다.
    def searchUrlByWord(self, words):
        wordlist=words.split(' ')
        wordlist=SearchEngine.__deleteStopwords(wordlist)
        
        Similarity=[]
        
        for i in range(len(self.wordsFreq)):     
            sim=0
            for word in wordlist:  
                for key, value in self.wordsFreq[i].items():
                    if word in key:
                        sim+=value    
            Similarity.append(sim)
        
        max_similarity=max(Similarity)
        for i in range(len(Similarity)):
            if max_similarity==Similarity[i]:
                print(self.urlList[i])
    
    
    
    #아래있는 메소드들은 이전에 만든 메소드 들이다.
    @staticmethod 
    def __deleteStyle(source):                      # <style> 제거 함수
        start=source.find('<style')
        end=source.find('</style>',start)
        check=True
        newstr=''
        if start>-1:
            for i in range(len(source)):
                if i==start:
                    check=False
                elif i==end:
                    check=True
                if check:
                    newstr+=source[i]
        else:
            return source

        return newstr
    
    @staticmethod
    def __deleteScript(source):                       #<script> 제거 함수
        start=source.find('<script')
        end=source.find('</script>',start)
        check=True
        newstr=''
        if start>-1:
            for i in range(len(source)):
                if i==start:
                    check=False
                elif i==end:
                    check=True
                if check:
                    newstr+=source[i]
        else:
            return source

        return newstr
    
    @staticmethod
    def __deleteStopwords(source):                       #불용어 제거 함수
        f= open('stopwords.txt','r',encoding='UTF-8')
        stopwords=f.readlines()
        f.close()
        for i in stopwords:
            while i in source:
                source.remove(i) 
        return source
    
    @staticmethod   
    def __wordsFrequency(url):                #단어의 빈도수를 사전객체로 반환하는 함수
        req = requests.get(url)
        source = req.text

        while(True):                        # <script> 처리
            newSource=SearchEngine.__deleteScript(source)
            if(newSource==source): 
                break
            source=newSource

        while(True):                        # <style> 처리
            newSource=SearchEngine.__deleteStyle(source)
            if(newSource==source): 
                break
            source=newSource


        word = ''                            # <   > 처리
        i=0

        while i<len(source):
            if source[i]=='<':
                while source[i]!='>':
                    i+=1
                i+=1
            else:
                word+=source[i]
                i+=1


        wordlist=word.split()                
        for i in range (len(wordlist)):       
            wordlist[i]=wordlist[i].strip(string.punctuation)    #구두 문자 처리

        wordlist=SearchEngine.__deleteStopwords(wordlist)   # 불용어 처리

        wordlist=[i for i in wordlist if i]  # '' <<비어있는 문자 처리    

        word_dic={}
        for key in wordlist:
            word_dic[key]=wordlist.count(key) # 빈도수를 통해 사전화 

        return word_dic
    
```

#### 2. 위 1번 문제에서 정의한 SearchEngine 클래스를 상속하여 SearchEngineWithOrderedWebWords 클래스를 정의하고 슈퍼클래스에 정의된 getWordsFrequency() 메소드를 오버라이드하여 단어 출현 빈도를 내림 차순으로 정렬하여 리스트로 출력하시오.

```python
class SearchEngineWithOrderedWebWords(SearchEngine):        
    
    #반복을 위한 __iter__ 이다.
    def __iter__(self):
        self.current = 0
        return self
    
    #단어의 빈도수를 정렬해주는 메소드이다.
    #따로 sorted_dict라는 멤버변수를 sorted메소드로 초기화 해주었다.
    #sorted메소드를 사용하여 튜플을 멤버로 갖는 리스트로 값을 리턴해준다.
    def getWordsFrequency(self, reverse=True):
        self.sorted_dict = sorted(self.totalFreq.items(), key = lambda item: item[1], reverse = reverse)
        return self.sorted_dict
    
    #반복을 위한 __next__ 이다.
    #self.current를 통해 다음값을 반환해준다. 위에서 정렬한 sorted_dict를 활용해 반환을 해준다.
    def __next__(self):
        if self.current<len(self.sorted_dict):
            self.current+=1
            return self.sorted_dict[self.current-1]
        else:
            raise StopIteration
```

```python
w2 = SearchEngineWithOrderedWebWords('http://www.times.com')
```

```python
w2.getWordsFrequency()
```

```python
w2.getWordsFrequency(reverse=False)
```

SearchEngine을 상속하고 getWordsFrequency 메소드를 재정의 하였다.  
reverse값을 주냐 안주냐에 따라 오름차순 내림차순으로 원하는 결과를 얻을 수 있다.

#### 3. 다음과 같은 코딩이 가능하도록 SearchEngineWithOrderedWebWords안에 반복자와 관련된 메소드를 추가하시오

iter와 next를 정의했기 때문에 객체를 for문에 넣어서 반복이 가능해진 모습이다.
